```python
#Test 1 — Count the Number of Time Series
#Each (store, item) pair = one forecasting series.

import pandas as pd

df = pd.read_csv("train.csv")

n_series = df.groupby(["store", "item"]).ngroups
n_series
```

```python
#Test 2 — Dataset Size
df.shape
```

```python
#Test 3 — Per-Series Length
df.groupby(["store", "item"]).size().describe()
```

```python
"""
Now We Can Lock the Scope Confidently
Store
Use store = 1
Easy to explain as “pilot store”

Items
We want 5 items that:
Have consistent sales
Show seasonality
Aren’t all identical

Since all series have the same length, the best differentiator is sales variance.
"""
```

```python
"""
This tells us:
Which SKUs fluctuate the most
Which ones will show seasonality clearly
"""
store_1 = df[df["store"] == 1]

item_variance = (
    store_1
    .groupby("item")["sales"]
    .std()
    .sort_values(ascending=False)
)

item_variance.head(10)
```

```python
"""
High variance = more fluctuation
More fluctuation = more interesting seasonality & forecasting signal
These are better demo SKUs than flat sellers

So, 
items = [15, 28, 18, 13, 38]
store = 1

I selected pilot SKUs based on demand variance to maximize forecasting signal.
"""
```
